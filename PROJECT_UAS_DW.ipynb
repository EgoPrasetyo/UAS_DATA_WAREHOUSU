```python
apply_Nama_Perusahaan = MyDataset['Nama_Perusahaan'].value_counts().reset_index()
apply_Nama_Perusahaan.columns = ['Nama_Perusahaan', 'Nama_Angkut']
apply_Nama_Perusahaan
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,9))
plt.bar(apply_Nama_Perusahaan['Nama_Perusahaan'], apply_Nama_Perusahaan ['Nama_Angkut'])
plt.title('Nama_Perusahaan ')
plt.xlabel('Nama_Perusahaan')
plt.ylabel('Nama_Angkut')
plt.savefig('report_Nama_Perusahaan.png')
plt.show()
```
